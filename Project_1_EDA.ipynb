```python
# Accessing the uploaded code

import pandas as pd
data = pd.read_csv('/content/all_data.csv')
data.head()
```

```python
# Converting `Low_Confidence_Limit` and `High_Confidence_Limit` columns to numeric, forcing errors to NaN
data['Low_Confidence_Limit'] = pd.to_numeric(data['Low_Confidence_Limit'], errors='coerce')
data['High_Confidence_Limit'] = pd.to_numeric(data['High_Confidence_Limit'], errors='coerce')

# Checking the conversion effect
conversion_check = data[['Low_Confidence_Limit', 'High_Confidence_Limit']].dtypes

# Filling missing values in `Data_Value` with the mean of the column (only if needed for numerical analysis)
data['Data_Value'].fillna(data['Data_Value'].mean(), inplace=True)

# Display the result of data type conversion and handling of missing values
conversion_check
```

```python
# Checking for basic information about the dataset
data_info = data.info()

# Summary statistics for numerical columns
summary_stats = data.describe(include='all')

# Checking for missing values in each column
missing_values = data.isnull().sum()

# Displaying the results
data_info, summary_stats, missing_values
```
